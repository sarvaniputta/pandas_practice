# For Baseball-specific analysis:
 

B) List the top 30 players in career success for stealing the most bases.
Use total of “Stolen Bases” + total of “Caught Stealing” and call that result "Stealing Attempts". Also compute and show their base stealing success rate as a percentage of attempts. In the results, show their name & the calculated numbers. (Join People and Batting tables)

```python
import pandas as pd
import numpy as np
import doctest
```

```python
people = pd.read_csv('baseballdatabank-2019.2/core/People.csv')
```

```python
#Checking if I imported it correctly or not

people.head()
```

```python
batting = pd.read_csv('baseballdatabank-2019.2/core/Batting.csv')
```

```python
#Checking if I imported it correctly

batting.head()
```

```python
# Now I will merge the two dataframes to solve the question


new_df = people.merge(batting, on = 'playerID')[['playerID', 'nameFirst', 'nameLast','nameGiven' ,'SB', 'CS']]


len(new_df)
```

```python
len(batting)
```

As the length of new_df and the batting dataframe are equal, I can say that there are no duplicate rows and the dataframes are joined correctly.

## Listing the top 30 players in career success for stealing the most bases. 

```python
new_df = new_df.groupby('playerID').agg({'SB':'sum','CS':'sum'}).sort_values(by = 'SB', ascending=False)
```

```python
new_df = new_df.head(30)
```

## Use total of “Stolen Bases” + total of “Caught Stealing” and call that result "Stealing Attempts". 

```python
new_df['Stealing Attempts'] = new_df['SB'] + new_df['CS']
```

```python
new_df.head()
```

```python
# Now I will calculate the base success stealing rate as percentage of attempts.
# Professor also informed about a formula to calculate so in one of the forum posts where the formula is:

# SB / (SB + CS) * 100

# Applying the above formula


new_df['% Base Success Stealing Rate'] = ((new_df['SB']/(new_df['SB'] + new_df['CS']))*100).round(2)


```

```python
new_df
```

### In the final answer, I am going to show their names and the calculated output

```python
answer_1 = new_df.merge(people, on = 'playerID')[['playerID', 'nameFirst', 'nameLast','nameGiven' ,'SB', 'CS','Stealing Attempts','% Base Success Stealing Rate']]
```
