### B) List the top 30 players in career success for stealing the most bases.
Use total of “Stolen Bases” + total of “Caught Stealing” and call that result "Stealing Attempts". Also compute and show their base stealing success rate as a percentage of attempts. In the results, show their name & the calculated numbers. 

(Join People and Batting tables)

```python
import pandas as pd
import numpy as np
```

#### I learnt a very useful function that is (.shift()[1:]). this function is very useful to shift the index number and start it from the desired position. 

I learnt this function from the third comment in the below link: 

https://stackoverflow.com/questions/20167930/start-index-at-1-for-pandas-dataframe 


When I was generating the output for my analysis, I observed that the index number started from zero but I wanted to start it from one.
That's why I looked for the mentioned concern and found the answer in the mentioned link.

```python
people = pd.read_csv('baseballdatabank-2019.2/core/People.csv').shift()[1:]
```

```python
#Checking if I imported it correctly or not

people.head()
```

```python
batting = pd.read_csv('baseballdatabank-2019.2/core/Batting.csv').shift()[1:]
```

```python
#Checking if I imported it correctly

batting.head()
```

```python
# Now I will merge the two dataframes to solve the question


new_df = people.merge(batting, on = 'playerID')[['playerID', 'nameFirst', 'nameLast','nameGiven' ,'SB', 'CS']]


len(new_df)
```

## Listing the top 30 players in career success for stealing the most bases. 

```python
new_df = new_df.groupby('playerID').agg({'SB':'sum','CS':'sum'}).sort_values(by = 'SB', ascending=False)
```

```python
# I am writing 31 instead of 30 because I have started the indexing from one using the shift function.
# head(31) will give me the top 30 players

new_df = new_df.head(31)

```

## Use total of “Stolen Bases” + total of “Caught Stealing” and call that result "Stealing Attempts". 

```python
new_df['Stealing Attempts'] = new_df['SB'] + new_df['CS']
```

```python
new_df.head()
```

```python
# Now I will calculate the base success stealing rate as percentage of attempts.
# Professor also informed about a formula to calculate so in one of the forum posts where the formula is:

# SB / (SB + CS) * 100

# Applying the above formula


new_df['% Base Success Stealing Rate'] = ((new_df['SB']/(new_df['SB'] + new_df['CS']))*100).round(2)


```

```python
new_df
```

### In the final answer, I am going to show their names and the calculated output

```python
top_30_players = new_df.merge(people, on = 'playerID')[['playerID', 'nameFirst', 'nameLast','nameGiven' ,'SB', 'CS','Stealing Attempts','% Base Success Stealing Rate']].shift()[1:]
```

```python
top_30_players
```

### E) List the top schools by number of alumni who were actually inducted into the Hall of Fame.

(Join Schools, CollegePlaying, HallOfFame)


```python
# To solve the above problem, I will first join the CollegePlaying file and HallOfFame file and
# then I'll add the resultant file with the Schools file.
```

```python
hall_of_fame = pd.read_csv('baseballdatabank-2019.2/core/HallOfFame.csv')
```

```python
college_playing = pd.read_csv('baseballdatabank-2019.2/core/CollegePlaying.csv')
```

```python
# I will need schools dataframe only to display the name of the school associated with a school ID. 

schools = pd.read_csv('baseballdatabank-2019.2/core/Schools.csv')
```

### Joining college_playing and hall_of_fame

```python
new_df2 = hall_of_fame.merge(college_playing, on = 'playerID').shift()[1:]
```

```python
new_df2.head()
```

```python
# Now I am only going to select the players, who were inducted.

new_df2 = new_df2[new_df2['inducted'] == 'Y']
new_df2 = new_df2[['schoolID', 'inducted']]
```

```python
# Finding total number of records who were inducted.

len(new_df2)
```

### I'll now group the records by schoolID and then count which school got the highest number of awards.

```python
new_df2 = new_df2.groupby('schoolID').count().sort_values(by = 'inducted', ascending=False)
```

```python
#Checking the answer

new_df2.head()
```

### Now I will join the 'schools' in order to display the name of the schools. 

```python
topschools = new_df2.merge(schools, on = 'schoolID')[['name_full', 'city', 'state', 'country', 'inducted']]

# I learnt the below function to rename column names from the below link:
# https://stackoverflow.com/questions/11346283/renaming-columns-in-pandas

topschools = topschools.rename(columns={'name_full': 'Name of School', 'city': 'City', 'state': 'State', 'country' : 'Country', 'inducted': 'Alumni in Hall of Fame'}).shift()[1:]
```
